Energ√≠a y Potencia

```python
import matplotlib.pyplot as plt
import numpy as np
import sympy as sym
from sympy import *

w = 300
L = 0.01
R = 4
Imax = 2
T0 = 2*np.pi/w
N = 100

t = np.linspace(0, T0, N)
i = Imax*np.sin(w*t)
plt.plot(t,i)
```

```python
Vmax = np.sqrt((R**2)+((w*L)**2))*Imax
ang = np.arctan(w*L/R)

v = Vmax*np.sin(w*t + ang)

plt.plot(t, i, label='I(t)')
plt.plot(t, v, label='V(t)')
plt.grid()
plt.legend()
plt.xlabel('Tiempo')
plt.ylabel('Amplitud')
```

```python
pi = v*i
pm = np.trapz(pi, t)/T0

plt.plot(t, i, label='I(t)')
plt.plot(t, v, label='V(t)')
plt.plot(t, pi, label='Pi(t)')
plt.plot(t, pm*np.ones(len(t)), label='Pm(t)')
plt.grid()
plt.legend()
plt.xlabel('Tiempo')
plt.ylabel('Amplitud')
```

```python
C = 0.01
Vmax = np.sqrt((R**2)+(1/((w*C)**2)))*Imax
ang = -np.arctan(1/(w*C*R))

v = Vmax*np.sin(w*t + ang)
```

```python
plt.plot(t, i, label='I(t)')
plt.plot(t, v, label='V(t)')
plt.grid()
plt.legend()
plt.xlabel('Tiempo')
plt.ylabel('Amplitud')
```

```python
pi = v*i
pm = np.trapz(pi, t)/T0

plt.plot(t, i, label='I(t)')
plt.plot(t, v, label='V(t)')
plt.plot(t, pi, label='Pi(t)')
plt.plot(t, pm*np.ones(len(t)), label='Pm(t)')
plt.grid()
plt.legend()
plt.xlabel('Tiempo')
plt.ylabel('Amplitud')
```
